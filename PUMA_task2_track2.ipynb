```python
# ================================================
# Batch Processing & Setup
# ================================================

import os
import glob
import json
import numpy as np
from tifffile import imread
from shapely.geometry import Polygon
from skimage.draw import polygon
import torch
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms
import matplotlib.pyplot as plt

# ------- Create a function to generate masks -------
def create_nuclei_mask(image_path, annotation_path):
    with open(annotation_path, 'r') as f:
        annotation_data = json.load(f)
    features = annotation_data['features']

    img = imread(image_path)
    if img.ndim == 3:
        height, width, _ = img.shape
    else:
        height, width = img.shape

    mask = np.zeros((height, width), dtype=np.uint8)
    for feat in features:
        props = feat.get('properties', {})
        geometry = feat.get('geometry', {})
        cls_id = get_nucleus_class_int(props)
        if geometry.get('type') == 'Polygon':
            coords = geometry['coordinates'][0]
            x_coords = [c[0] for c in coords]
            y_coords = [c[1] for c in coords]
            rr, cc = polygon(y_coords, x_coords, shape=(height, width))
            mask[rr, cc] = cls_id

    return img, mask

# ------- Batch Process All Training Data -------
annotation_files = glob.glob('./data/nuclei_annotations/01_training_dataset_geojson_nuclei/*.geojson')
output_image_dir = './data/processed_images'
output_mask_dir = './data/processed_masks'
os.makedirs(output_image_dir, exist_ok=True)
os.makedirs(output_mask_dir, exist_ok=True)
```

```python
for ann_file in annotation_files:
    base_name = os.path.basename(ann_file).replace('_nuclei.geojson', '')
    # Find matching image
    img_files = glob.glob(f'./data/ROIs/01_training_dataset_tif_ROIs/{base_name}.tif')
    if len(img_files) != 1:
        continue
    img_path = img_files[0]
    img, mask = create_nuclei_mask(img_path, ann_file)

    # Save npy files for convenience
    np.save(os.path.join(output_image_dir, base_name+'.npy'), img)
    np.save(os.path.join(output_mask_dir, base_name+'_mask.npy'), mask)

# At this point, I have .npy files for images and masks of all ROIs.
```

```python
# ------- Create a PyTorch Dataset for Training -------
class NucleiDataset(Dataset):
    def __init__(self, img_dir, mask_dir, transform=None):
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.transform = transform
        self.img_files = sorted(glob.glob(os.path.join(img_dir, '*.npy')))

    def __len__(self):
        return len(self.img_files)

    def __getitem__(self, idx):
        img_file = self.img_files[idx]
        base_name = os.path.splitext(os.path.basename(img_file))[0]
        mask_file = os.path.join(self.mask_dir, base_name+'_mask.npy')

        img = np.load(img_file)
        mask = np.load(mask_file)

        # If the image has RGBA, you can either use all channels or just RGB
        if img.ndim == 3 and img.shape[2] == 4:
            img = img[..., :3]

        # Convert to tensors
        img_tensor = torch.from_numpy(img.transpose(2,0,1)).float()  # C,H,W
        mask_tensor = torch.from_numpy(mask).long()  # H,W

        # Optional transforms (normalization, augmentation) can be applied here
        # E.g., simple normalization:
        img_tensor = img_tensor / 255.0

        return img_tensor, mask_tensor
```

```python
#Data agumentation
import os
import glob
import numpy as np
import torch
import cv2
from tifffile import imread
from skimage.draw import polygon
import albumentations as A
import albumentations.pytorch

###########################################
# PRE-EXISTING DIRECTORIES WITH ORIGINAL .NPY FILES
###########################################
original_image_dir = './data/processed_images'    # original .npy images
original_mask_dir = './data/processed_masks'      # original .npy masks

###########################################
# AUGMENTATION PARAMETERS
###########################################
target_classes = [3, 5, 6]  # Example of rare classes to focus on
num_samples = 10            # how many augmented samples to create per image containing target classes

# Define the offline augmentation transforms (similar to teammate's code)
offline_transforms = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.ShiftScaleRotate(shift_limit=0.05, scale_limit=0.1, rotate_limit=20, p=0.7),
    A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
])

###########################################
# CREATE DIRECTORIES FOR AUGMENTED DATA
###########################################
augmented_image_dir = './data/augmented_images'
augmented_mask_dir = './data/augmented_masks'
os.makedirs(augmented_image_dir, exist_ok=True)
os.makedirs(augmented_mask_dir, exist_ok=True)
```

```python
###########################################
# OFFLINE AUGMENTATION PROCESS
###########################################
img_files = sorted(glob.glob(os.path.join(original_image_dir, '*.npy')))

for img_file in img_files:
    base_name = os.path.splitext(os.path.basename(img_file))[0]
    mask_file = os.path.join(original_mask_dir, base_name+'_mask.npy')

    if not os.path.exists(mask_file):
        continue

    img = np.load(img_file)   # (H,W,C), float [0,255]?
    mask = np.load(mask_file) # (H,W)

    # Ensure img is uint8 for albumentations if needed
    # If originally saved as float [0-1], convert back to [0,255]
    # Check original saving logic: if you saved originally as [0,1], then do img = (img*255).astype(np.uint8)
    if img.dtype != np.uint8:
        img = (img * 255).astype(np.uint8)

    # Check if mask contains any target classes
    unique_classes = np.unique(mask)
    if not any(cls in unique_classes for cls in target_classes):
        # If no target classes, we skip augmenting this image
        continue

    # For each image containing target classes, create `num_samples` augmented variants
    for i in range(num_samples):
        augmented = offline_transforms(image=img, mask=mask)
        aug_img, aug_mask = augmented['image'], augmented['mask']

        # Save augmented samples as .npy
        aug_base_name = f"{base_name}_aug_{i}"
        # Convert aug_img back to [0,1] float if your training code expects that:
        aug_img = aug_img.astype(np.float32)/255.0

        np.save(os.path.join(augmented_image_dir, aug_base_name+'.npy'), aug_img)
        np.save(os.path.join(augmented_mask_dir, aug_base_name+'_mask.npy'), aug_mask)

print("Offline augmentation completed. Augmented samples are saved.")
```

```python
###########################################
# COMBINE ORIGINAL AND AUGMENTED DATASETS
###########################################
# After augmentation, we have original and augmented samples.
# We can just consider them together as one large dataset by merging directories or loading both sets.

# Let's merge directories logically: we have original and augmented .npy files now.
# You can either move augmented files into the original directories or just treat them together:
final_image_dir = './data/final_images'
final_mask_dir = './data/final_masks'
os.makedirs(final_image_dir, exist_ok=True)
os.makedirs(final_mask_dir, exist_ok=True)

# Copy original images
for f in glob.glob(os.path.join(original_image_dir, '*.npy')):
    base_name = os.path.splitext(os.path.basename(f))[0]
    if base_name.endswith('_mask'):
        # It's actually rare you'd have '_mask' in image directory, but if so:
        os.system(f'cp {f} {os.path.join(final_mask_dir, os.path.basename(f))}')
    else:
        os.system(f'cp {f} {os.path.join(final_image_dir, os.path.basename(f))}')

# Copy original masks
for f in glob.glob(os.path.join(original_mask_dir, '*.npy')):
    # All these should be masks
    os.system(f'cp {f} {os.path.join(final_mask_dir, os.path.basename(f))}')

# Copy augmented images
for f in glob.glob(os.path.join(augmented_image_dir, '*.npy')):
    base_name = os.path.splitext(os.path.basename(f))[0]
    if base_name.endswith('_mask'):
        # If any mask ended here by mistake, just in case:
        os.system(f'cp {f} {os.path.join(final_mask_dir, os.path.basename(f))}')
    else:
        os.system(f'cp {f} {os.path.join(final_image_dir, os.path.basename(f))}')

# Copy augmented masks
for f in glob.glob(os.path.join(augmented_mask_dir, '*.npy')):
    os.system(f'cp {f} {os.path.join(final_mask_dir, os.path.basename(f))}')

print("Merged original and augmented data into final_images and final_masks directories.")
```

```python
###########################################
# DATASET CLASS AND SPLITTING
###########################################
# Now define a dataset class for final data (no augmentation here, or minimal)
# We'll do on-the-fly minimal transforms if needed.
# For now, just normalization inside __getitem__.

class FinalNucleiDataset(torch.utils.data.Dataset):
    def __init__(self, img_dir, mask_dir):
        self.img_dir = img_dir
        self.mask_dir = mask_dir
        self.img_files = sorted(glob.glob(os.path.join(img_dir, '*.npy')))

    def __len__(self):
        return len(self.img_files)

    def __getitem__(self, idx):
        img_file = self.img_files[idx]
        base_name = os.path.splitext(os.path.basename(img_file))[0]
        mask_file = os.path.join(self.mask_dir, base_name+'_mask.npy')

        img = np.load(img_file)   # (H,W,C) float32 [0,1] if previously saved normalized or (H,W,C) uint8 [0,255] if not
        mask = np.load(mask_file) # (H,W)

        # If img is not normalized yet (e.g., if you saved after augmentation as [0,255]),
        # convert it to float32 and normalize:
        if img.dtype != np.float32:
            img = img.astype(np.float32) / 255.0

        # If the image has an alpha channel, drop it
        if img.ndim == 3 and img.shape[2] == 4:
            img = img[..., :3]

        # Convert to tensors
        # img: (H,W,C) -> (C,H,W)
        img_tensor = torch.from_numpy(img.transpose(2,0,1)).float()
        mask_tensor = torch.from_numpy(mask).long()

        return img_tensor, mask_tensor


final_dataset = FinalNucleiDataset(final_image_dir, final_mask_dir)
```

```python
# We already computed train_meta, train_primary, val_meta, val_primary, test_meta, test_primary previously.
# Let's assume you have them as in your previous attempt.

def compute_class_distribution_by_type(dset, num_classes=11):
    # num_classes = 11 (0 background + 10 nuclei classes)
    # We'll separate counts into metastatic and primary
    metastatic_counts = np.zeros(num_classes, dtype=int)
    primary_counts = np.zeros(num_classes, dtype=int)

    # To separate samples, we need to know their base_name
    # We have access to dset.subset if it's a RandomSplit or we can store filenames in dataset.
    # Since FinalNucleiDataset doesn't store base_names directly, we can modify it to do so.

    # Let's assume we can get filenames from final_dataset. If not, we must store them.
    # We have final_dataset.img_files. The base_name used to get masks is also known.
    # We'll reconstruct logic similarly.

    for i in range(len(dset)):
        img_file = dset.dataset.img_files[dset.indices[i]] if hasattr(dset, 'indices') else dset.img_files[i]
        base_name = os.path.splitext(os.path.basename(img_file))[0]
        # Determine if metastatic or primary
        if 'metastatic' in base_name:
            sample_type = 'metastatic'
        else:
            sample_type = 'primary'

        _, mask = dset[i]
        mask_np = mask.numpy().ravel()
        unique, cnts = np.unique(mask_np, return_counts=True)
        for u, c in zip(unique, cnts):
            if sample_type == 'metastatic':
                metastatic_counts[u] += c
            else:
                primary_counts[u] += c

    return metastatic_counts, primary_counts

# Convert to percentage and focus on classes 1 to 10 (exclude background 0).
def to_percentage(met_counts, pri_counts):
    # Focus only on classes 1 to 10
    met_classes = met_counts[1:]  # classes 1 to 10
    pri_classes = pri_counts[1:]  # classes 1 to 10
    met_total = met_classes.sum()
    pri_total = pri_classes.sum()

    #total = met_classes.sum() + pri_classes.sum()
    # Now total is only the sum of classes 1 to 10 pixels.
    met_pct = (met_classes / met_total) * 100
    pri_pct = (pri_classes / pri_total) * 100
    return met_pct, pri_pct

# Compute distributions
train_meta, train_primary = compute_class_distribution_by_type(train_dataset, num_classes=11)
val_meta, val_primary = compute_class_distribution_by_type(val_dataset, num_classes=11)
test_meta, test_primary = compute_class_distribution_by_type(test_dataset, num_classes=11)

train_met_pct, train_pri_pct = to_percentage(train_meta, train_primary)
val_met_pct, val_pri_pct = to_percentage(val_meta, val_primary)
test_met_pct, test_pri_pct = to_percentage(test_meta, test_primary)

# We have 10 classes. We'll create stacked horizontal bars.
# We'll use the `colors` you defined earlier for classes 1 to 10:
# Ensure colors[1] to colors[10] are available, or slice colors[1:11]

fig, axes = plt.subplots(1, 3, figsize=(18,6))
datasets = [
    ("Training Set", train_met_pct, train_pri_pct),
    ("Validation Set", val_met_pct, val_pri_pct),
    ("Test Set", test_met_pct, test_pri_pct)
]

class_labels = [inv_class_map[i+1] for i in range(10)]  # get names from i+1 in class_map
class_colors = colors[1:11]  # classes 1 to 10 colors

y_labels = ["Metastatic Samples", "Primary Samples"]

for ax, (title, met_pct, pri_pct) in zip(axes, datasets):
    # We'll plot two horizontal bars:
    # top bar: metastatic
    # bottom bar: primary

    # Prepare cumulative sums for stacking
    met_cum = np.zeros_like(met_pct)
    pri_cum = np.zeros_like(pri_pct)

    # Y positions: 0 for metastatic, 1 for primary
    y_pos = [0, 1]

    # For each class, we add a segment
    for i in range(10):
        ax.barh(y_pos[0], met_pct[i], left=met_cum[i], color=class_colors[i], edgecolor='white')
        met_cum[i+1:] += met_pct[i]  # accumulate for next classes
        # Actually we should accumulate a separate variable:
    # Actually, we need separate cumulative arrays:
    met_left = 0
    pri_left = 0
    for i in range(10):
        # metastatic
        ax.barh(y_pos[0], met_pct[i], left=met_left, color=class_colors[i], edgecolor='white')
        met_left += met_pct[i]
        # primary
        ax.barh(y_pos[1], pri_pct[i], left=pri_left, color=class_colors[i], edgecolor='white')
        pri_left += pri_pct[i]

    ax.set_xlim(0,100)
    ax.set_yticks(y_pos)
    ax.set_yticklabels(y_labels)
    ax.set_xlabel('Percentage (%)')
    ax.set_title(title)

# Create legend for classes
legend_patches = []
for i in range(10):
    legend_patches.append(plt.Line2D([0],[0], marker='o', color='w', label=class_labels[i],
                                     markerfacecolor=class_colors[i], markersize=10))

axes[-1].legend(handles=legend_patches, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.tight_layout()
plt.show()
```

```python
##########################################
# Combined Loss (CE + Dice)
##########################################
import segmentation_models_pytorch as smp
import torch.nn as nn
import torch.nn.functional as F

class CombinedLoss(nn.Module):
    def __init__(self, num_classes, weight=None, alpha=0.5):
        super(CombinedLoss, self).__init__()
        self.num_classes = num_classes
        self.ce_loss = nn.CrossEntropyLoss(weight=weight)
        self.dice_loss = smp.losses.DiceLoss('multiclass', smooth=0.01, ignore_index=0)
        self.alpha = alpha

    def forward(self, outputs, targets):
        # CE loss
        ce_loss = self.ce_loss(outputs, targets)
        # Dice loss
        dice_loss = self.dice_loss(outputs, targets)
        combined_loss = self.alpha * ce_loss + (1 - self.alpha) * dice_loss
        return combined_loss

weights_tensor = None
criterion = CombinedLoss(num_classes=num_classes, weight=weights_tensor, alpha=0.5)
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
```

```python
##########################################
# Training with Early Stopping
##########################################
import numpy as np

num_epochs = 100
patience = 7
epochs_no_improve = 0
min_val_loss = np.inf
min_val_loss_epoch = 0
best_model_path = "best_hover_next_model.pth"
train_loss_history = []
val_loss_history = []


for epoch in range(num_epochs):
    model.train()
    train_loss = 0.0
    for imgs, masks in train_loader:
        imgs = imgs.cuda()
        masks = masks.cuda()
        optimizer.zero_grad()
        outputs = model(imgs)
        # Class channels start at index 1 (since index 0 is instance)
        class_channels = outputs[:, 1:1+num_classes, :, :]
        loss = criterion(class_channels, masks)
        train_loss += loss.item()
        loss.backward()
        optimizer.step()
    train_loss /= len(train_loader)

    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for imgs, masks in val_loader:
            imgs = imgs.cuda()
            masks = masks.cuda()
            outputs = model(imgs)
            class_channels = outputs[:, 1:1+num_classes, :, :]
            loss = criterion(class_channels, masks)
            val_loss += loss.item()
    val_loss /= len(val_loader)
    train_loss_history.append(train_loss)
    val_loss_history.append(val_loss)

    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Validation Loss: {val_loss:.4f}")

    if val_loss < min_val_loss:
        min_val_loss = val_loss
        min_val_loss_epoch = epoch + 1
        epochs_no_improve = 0
        torch.save({
            'epoch': epoch+1,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': val_loss,
        }, best_model_path)
        print("Best model saved!")
    else:
        epochs_no_improve += 1
        if epochs_no_improve >= patience:
            print(f"Early stopping triggered at epoch {epoch+1}")
            break

print(f"Training complete. Minimum Validation Loss: {min_val_loss} at Epoch {min_val_loss_epoch}")
# Plot training and validation loss over epochs
plt.figure(figsize=(8,6))
plt.plot(train_loss_history, label='Training Loss')
plt.plot(val_loss_history, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss over Epochs')
plt.legend()
plt.show()

```

```python
num_classes = 11
output_dir = "./predictions_json"
output_dir_gt = "./predictions_gt_json"

# We'll integrate the provided code. First, define the functions you pasted:

import os, json, cv2, torch
import numpy as np
from scipy.ndimage import label

def mask_to_polygons(mask: np.ndarray, num_classes: int):
    class_polygons = {}
    for cls in range(num_classes):
        binary_mask = (mask == cls).astype(np.uint8)
        contours, _ = cv2.findContours(binary_mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        polygons = [
            np.squeeze(contour, axis=1)
            for contour in contours
            if contour.shape[0] >= 3
        ]
        class_polygons[cls] = polygons
    return class_polygons

def generate_json_from_model(model, loader, num_classes, output_dir, output_dir_gt):
    model.eval()
    os.makedirs(output_dir, exist_ok=True)
    os.makedirs(output_dir_gt, exist_ok=True)
    with torch.no_grad():
        for batch_idx, (imgs, masks) in enumerate(loader):
            imgs = imgs.cuda()
            masks = masks.cuda()
            outputs = model(imgs)
            # Assuming classification channels come first: If not, adapt indexing
            # In your code, classification channels started at index 1, but
            # the teammate’s code seems to assume class channels start at index 0.
            # Adjust as needed. Here we assume class_pred = outputs[:,1:1+num_classes,:,:]
            class_pred = outputs[:, 1:1+num_classes, :, :]
            for i in range(class_pred.shape[0]):
                pred_classes = torch.argmax(class_pred[i], dim=0).cpu().numpy()  # (H,W)
                mask_np = masks[i].cpu().numpy()  # ground truth (H,W)

                polygon = mask_to_polygons(pred_classes, num_classes=num_classes)
                polygon_gt = mask_to_polygons(mask_np, num_classes=num_classes)

                json_data = {"type": "Multiple polygons", "polygons": []}
                json_data_gt = {"type": "Multiple polygons", "polygons": []}

                # Add predicted polygons
                for cls, polygons in polygon.items():
                    if cls != 0:
                        for poly in polygons:
                            seed_point = [float(poly[0][0]), float(poly[0][1]), 0.5]
                            path_points = [[float(x), float(y), 0.5] for x, y in poly]
                            polygon_data = {
                                "name": f"class_{cls}",
                                "seed_point": seed_point,
                                "path_points": path_points,
                                "sub_type": "",
                                "groups": [],
                                "probability": 1.0
                            }
                            json_data["polygons"].append(polygon_data)

                # Add ground truth polygons
                for cls, polygons in polygon_gt.items():
                    if cls != 0:
                        for poly in polygons:
                            seed_point = [float(poly[0][0]), float(poly[0][1]), 0.5]
                            path_points = [[float(x), float(y), 0.5] for x, y in poly]
                            polygon_data = {
                                "name": f"class_{cls}",
                                "seed_point": seed_point,
                                "path_points": path_points,
                                "sub_type": "",
                                "groups": [],
                                "probability": 1.0
                            }
                            json_data_gt["polygons"].append(polygon_data)

                pred_filename = os.path.join(output_dir, f"prediction_{batch_idx}_{i}.json")
                with open(pred_filename, 'w') as jf:
                    json.dump(json_data, jf, indent=4)

                gt_filename = os.path.join(output_dir_gt, f"gt_{batch_idx}_{i}.json")
                with open(gt_filename, 'w') as jf:
                    json.dump(json_data_gt, jf, indent=4)

    print(f"JSON files saved in: {output_dir} and {output_dir_gt}")

def calculate_centroid(points):
    points = np.array(points)
    centroid = np.mean(points, axis=0)
    return centroid

def extract_features_from_json(json_data, json_name):
    features_list = []
    for polygon_data in json_data.get('polygons', []):
        category = polygon_data['name']
        score = polygon_data.get('score', 1)
        path_points = polygon_data['path_points']

        if len(path_points) < 3:
            continue
        exterior_coords = [coord[:2] for coord in path_points]
        centroid = calculate_centroid(exterior_coords)
        features_list.append({
            'filename': json_name,
            'category': category,
            'centroid': centroid.tolist(),
            'score': score
        })
    return features_list

def process_json_file(json_file_path):
    json_name = os.path.basename(json_file_path)
    with open(json_file_path, 'r') as f:
        json_data = json.load(f)
        features = extract_features_from_json(json_data, json_name)
    return features

def calculate_centroid_distance(gt_features, pred_features):
    pred_structure = {}
    for pred_feature in pred_features:
        match_key = pred_feature['category']
        if match_key not in pred_structure:
            pred_structure[match_key] = []
        pred_structure[match_key].append(pred_feature)

    results = []
    for gt_feature in gt_features:
        match_key = gt_feature['category']
        eligible_predictions = []
        if match_key in pred_structure:
            for pred_feature in pred_structure[match_key]:
                distance = np.linalg.norm(np.array(gt_feature['centroid']) - np.array(pred_feature['centroid']))
                if distance < 15:
                    eligible_predictions.append({
                        'pred_json': pred_feature['filename'],
                        'gt_category': gt_feature['category'],
                        'pred_category': pred_feature['category'],
                        'distance': distance,
                        'pred_score': pred_feature['score'],
                        'pred_feature': pred_feature,
                    })
        eligible_predictions.sort(key=lambda x: (-x['pred_score'], x['distance']))
        if eligible_predictions:
            best_match = eligible_predictions[0]
            results.append(best_match)
            for i, pred in enumerate(pred_structure[match_key]):
                if np.array_equal(pred['centroid'], best_match['pred_feature']['centroid']):
                    del pred_structure[match_key][i]
                    break
    return results

def calculate_classification_metrics(results, gt_features, pred_features):
    pred_tp = [match['pred_category'] for match in results]
    ground_truth = [feature['category'] for feature in gt_features]
    pred_all = [feature['category'] for feature in pred_features]

    gt_dict = dict(zip(*np.unique(ground_truth, return_counts=True)))
    pred_dict = dict(zip(*np.unique(pred_all, return_counts=True)))
    tp_dict = dict(zip(*np.unique(pred_tp, return_counts=True)))

    micro_TP, micro_FP, micro_FN = 0, 0, 0
    results_metrics = {}

    all_categories = np.unique(list(gt_dict.keys()) + list(pred_dict.keys()))
    for category in all_categories:
        TP = tp_dict.get(category, 0)
        FP = pred_dict.get(category, 0) - TP
        FN = gt_dict.get(category, 0) - TP
        micro_TP += TP
        micro_FP += FP
        micro_FN += FN

        precision = TP / (TP + FP + 1e-8)
        recall = TP / (TP + FN + 1e-8)
        f1_score = 2 * precision * recall / (precision + recall + 1e-8)

        results_metrics[category] = {
            'TP': TP, 'FP': FP, 'FN': FN,
            'precision': precision, 'recall': recall, 'f1_score': f1_score
        }

    micro_precision = micro_TP / (micro_TP + micro_FP + 1e-8)
    micro_recall = micro_TP / (micro_TP + micro_FN + 1e-8)
    micro_f1_score = 2 * micro_precision * micro_recall / (micro_precision + micro_recall + 1e-8)

    macro_f1_score = np.mean([m['f1_score'] for m in results_metrics.values()])
    results_metrics['micro'] = {'precision': micro_precision, 'recall': micro_recall, 'f1_score': micro_f1_score}
    results_metrics['macro'] = {'f1_score': macro_f1_score}

    return results_metrics

def evaluate_files(ground_truth_file, pred_file):
    gt_features = process_json_file(ground_truth_file)
    pred_features = process_json_file(pred_file)
    results = calculate_centroid_distance(gt_features, pred_features)
    metrics = calculate_classification_metrics(results, gt_features, pred_features)
    return metrics

def evaluate_all_files(ground_truth_dir, pred_file_dir):
    overall_results = {}
    for gt_filename in os.listdir(ground_truth_dir):
        if not gt_filename.endswith('.json'):
            continue
        pred_filename = "prediction" + gt_filename[2:]
        gt_filepath = os.path.join(ground_truth_dir, gt_filename)
        pred_filepath = os.path.join(pred_file_dir, pred_filename)
        if not os.path.exists(pred_filepath):
            print(f"Warning: No prediction file found for {gt_filename}")
            continue
        try:
            file_metrics = evaluate_files(gt_filepath, pred_filepath)
            overall_results[gt_filename] = file_metrics
        except Exception as e:
            print(f"Error processing {gt_filename}: {e}")
    return overall_results
```
